
# Numerical Integration Methods

This notebook provides an overview of common numerical integration methods, including Trapezoidal Rule, Simpson's Rule, and Gaussian Quadrature. Each method includes a step-by-step algorithm and example Python code.



## Trapezoidal Rule

### Algorithm:
1. Divide the interval [a, b] into n subintervals of equal width h = (b - a)/n.
2. Compute the sum: Integral ≈ (h/2) * [f(x₀) + 2f(x₁) + 2f(x₂) + ... + 2f(xₙ₋₁) + f(xₙ)].
3. Return the result.


```python
import numpy as np
import matplotlib.pyplot as plt

a,b=0,1
n=10
x=np.linspace(a,b,n+1)



def trapaziod_rule(f,a,b,n):
  h=(b-a)/n
  result=f(a)+f(b)
  for i in range (1,n):
    result+=2*f(a+i*h)
  return result*(h/2)

def f(x):
  return np.exp(x)


y=f(x)
trap_integral=trapaziod_rule(f,a,b,n)
print(trap_integral)

plt.plot(x,y)
plt.show()



```


## Simpson's Rule

### Algorithm:
1. Divide the interval [a, b] into an even number n of subintervals.
2. Compute the sum: Integral ≈ (h/3) * [f(x₀) + 4f(x₁) + 2f(x₂) + 4f(x₃) + ... + 4f(xₙ₋₁) + f(xₙ)].
3. Return the result.


```python
a,b=0,1
n=10
x=np.linspace(a,b,n+1)



def simpson_rule(f,a,b,n):
  if n%2:
    raise ValueError("n must be even")
  h=(b-a)/n
  result=f(a)+f(b)
  for i in range (1,n):
    if i%2:
      result+=4*f(a+i*h)
    else:
      result+=2*f(a+i*h)
  return result*(h/3)


def f(x):
  return np.exp(x)


y=f(x)
simpson_integral=simpson_rule(f,a,b,n)
print(simpson_integral)
```
